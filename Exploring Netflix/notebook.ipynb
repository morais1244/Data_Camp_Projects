You work for a production company that specializes in nostalgic styles. You want to do some research on movies released in the 1990's. You'll delve into Netflix data and perform exploratory data analysis to better understand this awesome movie decade!

You have been supplied with the dataset `netflix_data.csv`, along with the following table detailing the column names and descriptions. Feel free to experiment further after submitting!

## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Filter the DataFrame to include only movies
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

# Further filter the DataFrame to include only movies released in the 1990s
movies_1990s = netflix_subset[(netflix_subset["release_year"] >= 1990) & (netflix_subset["release_year"] < 2000)]

# Plot a histogram of the duration of movies from the 1990s
plt.hist(movies_1990s["duration"])
plt.title('Distribuição da duração dos filmes na década de 90')  # Title in Portuguese: Distribution of movie durations in the 90s
plt.xlabel('Duração (minutos)')  # X-axis label in Portuguese: Duration (minutes)
plt.ylabel('Quantidade de filmes')  # Y-axis label in Portuguese: Number of movies
plt.show()

# Define a variable for duration
duration = 100

# Filter the DataFrame to include only action movies from the 1990s
action_movies = movies_1990s[movies_1990s['genre']=='Action']

# Count the number of action movies with a duration of less than 90 minutes
short_movie_count = (action_movies["duration"]<90).sum()

# Print the count of short action movies
print(short_movie_count)
```

## Análise dos Dados Coletados

### Introdução
Os dados coletados referem-se a filmes disponíveis na plataforma Netflix, com foco específico em filmes lançados na década de 1990. A análise inclui a distribuição da duração desses filmes e uma contagem específica de filmes de ação com duração inferior a 90 minutos.

### Distribuição da Duração dos Filmes na Década de 90
A distribuição da duração dos filmes lançados na década de 1990 foi visualizada através de um histograma. Abaixo estão os principais pontos observados:

- **Título do Gráfico:** Distribuição da duração dos filmes na década de 90
- **Eixo X:** Duração (minutos)
- **Eixo Y:** Quantidade de filmes

O histograma fornece uma visão clara de como as durações dos filmes estão distribuídas, permitindo identificar se há uma tendência para durações mais curtas ou mais longas.

### Filmes de Ação na Década de 90
Além da análise geral, foi realizada uma filtragem específica para identificar filmes de ação lançados na década de 1990. A análise focou em contar quantos desses filmes de ação têm uma duração inferior a 90 minutos.

- **Número de Filmes de Ação com Duração Inferior a 90 Minutos:** `short_movie_count`

Essa contagem é útil para entender a prevalência de filmes de ação mais curtos durante esse período, o que pode refletir tendências de produção e preferências do público na época.
