Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

````python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Sort the schools by average math score in descending order
schools_sort_math = schools.sort_values('average_math', ascending=False)

# Filter schools with average math scores greater than 80% of the maximum score (800)
best_schools_subset = schools_sort_math[schools_sort_math['average_math'] > (0.8 * 800)]

# Select only the school name and average math score columns
best_math_schools = best_schools_subset[['school_name', 'average_math']]

# Preview the best math schools
best_math_schools.head()

# Calculate the total SAT score by summing the average scores of math, reading, and writing
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']

# Sort the schools by total SAT score in descending order
schools_sort_sat = schools.sort_values('total_SAT', ascending=False)

# Select the top 10 schools based on total SAT score
top_10_schools = schools_sort_sat[:10][['school_name', 'total_SAT']]

# Display the top 10 schools
top_10_schools

# Group the schools by borough and calculate the count, mean, and standard deviation of total SAT scores
boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)

# Filter for the borough with the maximum standard deviation in total SAT scores
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]

# Rename the columns for clarity
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})

# Optional: Move borough from index to column
largest_std_dev.reset_index(inplace=True)
````
